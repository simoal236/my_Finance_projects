```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
paths = 5000       ## 'path' could stand for the number of stocks in a ptf
dt = 1/252
year = 1
obs = int(year/dt)

s0 = 100
mu = 0.05
sigma = 1.2

sT = np.zeros((obs +1,paths))    ## 'sT' represents all the possible future value of s
sT[0:,:] = s0                   ## the first row of the matrix, i.e. the value of all paths today, is equal to s0; it is the starting point
epsilon = np.random.randn(obs,paths) 
wT = sigma * np.sqrt(dt)*epsilon        ## Wiener process normally distributed
drift = (mu - 0.5*sigma*sigma) * np.ones((obs, 1)) * dt

sT[1:,:] = np.exp(drift + wT)
sT_paths = np.cumprod(sT, axis =0)
```

```python
plt.plot(sT_paths[:,10:20]);
```

```python
ts = np.zeros(obs+1)
ts [1:] = dt
ts = np.cumsum(ts)
```

```python
plt.plot(ts,sT_paths[:,:15]);
```
